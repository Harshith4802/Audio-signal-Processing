```python
import matplotlib.pyplot as plt
import numpy as np
import wave
from playsound import playsound
```

```python
#prev functions

#Analysys equation
def DTFT(signal,n,w):
    y = []
    #Appending the summations for all omegas(w)
    for i in range(len(w)):

        #Summation for analysis equation for a particular omega(w[i])
        sum = 0
        for j in range(len(signal)):
              sum += signal[j]*complex(np.cos(w[i]*n[j]),-np.sin(w[i]*n[j]))
        y.append(np.round(sum,decimals = 4))
    return y

def plot_spectrums(y,w):
    mag = []
    phase = []
    for i in range(len(w)):
        mag.append(abs(y[i]))
        p = np.angle(y[i])
        phase.append(p)
  #Plotting the spectrums
    plt.subplot(2,1,1)
    plt.title("Magnitude spectrum")
    plt.xlabel("ω")
    plt.ylabel("magnitude")
    plt.plot(w,mag)
    plt.show()
    plt.subplot(2,1,2)
    plt.title("Phase spectrum")
    plt.xlabel("ω")
    plt.ylabel("phase(radians)")
    plt.plot(w,phase)
    plt.show()
    
def group_delay(ph_h):
    grd_h = []
    for i in range(N-1):
        grd = -(ph_h[i+1]-ph_h[i-1])/(2*(w[i+1]-w[i]))
        grd_h.append(round(grd,4))
    grd_h.append(grd)
    return grd_h
```

```python
#Q1
n = np.linspace(-1,1,3)
h = [3,6,3]

#Number of values of w
N = 1000
#omega in the range -π ≤ ω < π
w = np.linspace(-np.pi,np.pi,N,endpoint = False)

dtft_h = DTFT(h,n,w)
plot_spectrums(dtft_h,w)
print("From the magnitude spectrum we can see that this is a low pass filter")
print("Since the phase spectrum is constant 0,we can say the system has zero-phase response.")
```

```python
#Q2(a)
l = 5
n1 = np.linspace(0,l-1,l)
h_type1 = np.array([3,2,1,2,3])
m = np.array([-1,-1,0,1,1])
h_type3 = h_type1*m
print("Type-3 = ",h_type3)
```

```python
#Q2(b)
l = 6
n2 = np.linspace(0,l-1,l)
h_type2 = np.array([3,2,1,1,2,3])
# m = np.array([-1,-1,-1,1,1,1])
# h_type4 = h_type2*m
h_type4= []
for i in n2:
    h_type4.append(h_type2[int(i)]*((-1)**i))
    
print("Type-4 = ",h_type4)
```

```python
#Q2(c)
#TYPE-1
print("Type-1:")
dtft_h_type1 = DTFT(h_type1,n1,w)
plot_spectrums(dtft_h_type1,w)
phase = []
mag1 = []
for i in range(len(w)):
    mag1.append(abs(dtft_h_type1[i]))
    phase.append(np.angle(dtft_h_type1[i]))
grd = group_delay(phase)
plt.xlabel("w")
plt.ylabel("grd(H_type1(e^jω))")
plt.plot(w,grd) 
plt.ylim((1,3))
plt.show()
```

```python
#TYPE-2
print("Type-2:")
dtft_h_type2 = DTFT(h_type2,n2,w)
plot_spectrums(dtft_h_type2,w)
phase = []
mag2 = []
for i in range(len(w)):
    mag2.append(abs(dtft_h_type2[i]))
    phase.append(np.angle(dtft_h_type2[i]))
grd = group_delay(phase)
plt.xlabel("w")
plt.ylabel("grd(H_type2(e^jω))")
plt.plot(w,grd) 
plt.ylim((1,3))
plt.show()
```

```python
#TYPE-3
print("Type-3:")
dtft_h_type3 = DTFT(h_type3,n1,w)
plot_spectrums(dtft_h_type3,w)
phase = []
mag3 = []
for i in range(len(w)):
    mag3.append(abs(dtft_h_type3[i]))
    phase.append(np.angle(dtft_h_type3[i]))
grd = group_delay(phase)
plt.xlabel("w")
plt.ylabel("grd(H_type3(e^jω))")
plt.plot(w,grd) 
plt.ylim((1,3))
plt.show()
```

```python
#TYPE-4
print("Type-4:")
dtft_h_type4 = DTFT(h_type4,n2,w)
plot_spectrums(dtft_h_type4,w)
phase = []
mag4 = []
for i in range(len(w)):
    mag4.append(abs(dtft_h_type4[i]))
    phase.append(np.angle(dtft_h_type4[i]))
grd = group_delay(phase)
plt.xlabel("w")
plt.ylabel("grd(H_type4(e^jω))")
plt.plot(w,grd) 
plt.ylim((1,3))
plt.show()

print("We can see that all the four systems have constant group delay except at discontinuities in the phase response")
print("Group delay for Type-1 and Type-3 systems are same(2)")
print("Group delay for Type-2 and Type-4 systems are same(2.5)")
```
